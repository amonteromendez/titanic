- Get data sample from file:

```python
file = 'data/train.csv' #file path
n = 3                   #number of rows to read
with open(file,'r') as file:
    for _ in range(n):
        print(file.readline(), end='\r')
```

- Get the number of columns on the file (columns count):

```python
file = 'data/train.csv' #file path
with open(file,'r') as file:
    print(len(file.readline().split(','))) #',' separated file
```

### Reading a file using Pandas

- Reading a local file:

```python
path = 'data/train.csv'
df = pd.read_csv(path)
df.head()
```

- Reading a file from a web url:

```python
path = 'https://github.com/stefmolin/Hands-On-Data-Analysis-with-Pandas-2nd-edition/blob/master/ch_02/data/earthquakes.csv?raw=True'
df = pd.read_csv(path)
df.head()
```

### Writing to a file using Pandas

```python
path = 'data/output_example.csv'
df.to_csv(path, index = False)
```
